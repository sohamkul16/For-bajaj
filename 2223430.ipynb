```python
import pandas as pd

df = pd.read_csv('/Book1.csv')

df.sort_values(by=["student_id", "attendance_date"])

latest_absence_streaks = []

for student_id, group in df.groupby("student_id"):
    group = group.reset_index(drop=True)

    streaks = []
    start_date = None
    count = 0

    for i, row in group.iterrows():
        if row["status"] == "Absent":
            if count == 0:
                start_date = row["attendance_date"]
            count += 1
        else:
            if count > 3:
                streaks.append((start_date, group.loc[i - 1, "attendance_date"], count))
            count = 0

    if count > 3:
        streaks.append((start_date, group.iloc[-1]["attendance_date"], count))

    if streaks:
        latest_streak = streaks[-1]
        latest_absence_streaks.append([student_id, latest_streak[0], latest_streak[1], latest_streak[2]])

absence_df = pd.DataFrame(latest_absence_streaks, columns=["student_id", "absence_start_date", "absence_end_date", "total_absent_days"])

print(absence_df)
```
